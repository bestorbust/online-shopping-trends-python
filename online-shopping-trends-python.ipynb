```python
category_sales = df.groupby('Product_Category')['Quantity'].sum().sort_values(ascending=True)
```

 Here we calculate the total quantity of products sold for each product category in a Dataset. It groups the data by the **Product_Category** column, sums up the **quantities** for each category, and then sorts the results in **ascending order** of total quantity. The variable **category_sales** holds this information.

## Which is the highest selling product?

most of our guess would be accessories or makeup.Let's see is our guess is true or not...

```python
plt.figure(figsize=(12,5))
plt.title('Highest Selling Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Total Sales Quantity')
plt.xticks(rotation=45) 
plt.bar(category_sales.index, category_sales.values)
plt.show()
```

When it comes to shopping online most of our guess would be clothes or accessories.considering how popular these items usually are in online shopping.

However,the data revealed a different trend which is office related product . Highest selling product is office related products and not what we expected. Suprisingly,makeup/accessories doesn't top the list. The second most purchased items online is apparel.

This unexpected outcome is likely influenced by the COVID-19 pandemic, as work from home has increased the demand for office supplies. The shift in lifestyle, with more time spent at home, also explains the increased sales in categories like 'Drinkware' and 'Lifestyle.'

### What is the most popular month to shop online?
lets find out the answer for this..!


```python
monthly_sales = df['Month'].value_counts()
```

  **monthly_sales** will  contain the count of sales transactions for each month.

```python
plt.figure(figsize=(10, 6))
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.title('Monthly Online Shopping Activity')
plt.xticks(range(1, 13),['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.bar(monthly_sales.index, monthly_sales.values)

plt.show()

```

Most popular month for online shopping is August.The increase in sales during August can be explained by various factors such as back-to-school shopping, end-of-summer sales, and people getting ready for upcoming holidays.

### What is most popular location in United States to Online Shop in?

 lets find the answer for this by analysising the data.


```python
location_popularity = df['Location'].value_counts().reset_index()
location_popularity.columns = ['Location', 'Number of Shoppers']
```

  **location_popularity** is a DataFrame that shows the popularity of different locations based on the number of shoppers, and it has two columns: 'Location' and 'Number of Shoppers'.

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=location_popularity, y='Number of Shoppers', x='Location')
plt.title('Most Popular Locations for Online Shopping in the United States')
plt.ylabel('Number of Shoppers')
plt.xlabel('Location')
plt.show()
```

**Chicago** is the top location for online shopping in the dataset. This suggests that people living in Chicago and the surrounding areas shop online more often than those in other parts of the United States. The trend might be influenced by the harsh winter weather in Chicago, with extremely cold temperatures.
